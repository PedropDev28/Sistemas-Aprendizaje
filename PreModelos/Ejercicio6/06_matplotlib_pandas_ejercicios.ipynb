1. Utilizando el fichero *democanarias.csv*. Muestra este gráfico de barras de número de habitantes por isla: 




```python
import matplotlib.pyplot as plt
import pandas as pd


# Leer el archivo CSV
csv = pd.read_csv(
    "democanarias.csv", delimiter=","
)  # Asegúrate de usar coma como delimitador
df = pd.DataFrame(csv)

fig, ax = plt.subplots()

# Ver las cabeceras de las columnas
print(df.columns)

# Crear el gráfico
plt.title("Población de las islas", fontsize=16)
plt.xlabel("Islas")

plt.grid()

# Graficar las barras
plt.bar(df["Isla"], df["Poblacion"], color="blue")

# Rotar las etiquetas del eje X (nombres de las islas) de forma vertical
plt.xticks(rotation=90)

plt.legend(["Población"], loc="upper right")

ax.set_ylabel(rotation=0, labelpad=20, ylabel="Habitantes")

plt.show()
```

Escribe un programa que divida la ventana en tres partes. 

Arriba a la izquierda, pinte un diagrama de barras en rojo con los gastos de un año, agrupado en meses.

 En la derecha, pinte un diagrama de barras en azul, con los ingresos en meses
 
Y abajo, un diagrama de líneas con con los ingresos frente a los gastos, a lo largo de los meses

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

# Datos
meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']

ingresos = [100,300,150,400,150,250,300,300,320,340,55,90]

gastos = [90,150,100,100,120,150,200,100,20,40,55,80]

# Crear un layout con gridspec
fig = plt.figure(figsize=(8, 10))
gs = gridspec.GridSpec(3, 2, height_ratios=[1, 1, 2])

# Gráfico de barras de Gastos
ax1 = fig.add_subplot(gs[0, 0])
ax1.bar(meses, gastos, color='red')
ax1.set_title('Gastos')
ax1.tick_params(axis='x', rotation=45)

# Gráfico de barras de Ingresos
ax2 = fig.add_subplot(gs[0, 1])
ax2.bar(meses, ingresos, color='blue')
ax2.set_title('Ingresos')
ax2.tick_params(axis='x', rotation=45)

# Gráfico de líneas combinadas
ax3 = fig.add_subplot(gs[1:, :])  # Ocupa dos filas y dos columnas
ax3.plot(meses, ingresos, '-o', label='Ingresos', color='red', marker='v')
ax3.plot(meses, gastos, '-o', label='Gastos', color='blue', marker='X')
ax3.set_title('Ingresos frente a gastos')
ax3.set_ylabel('Miles de € (k)')
ax3.legend()
ax3.tick_params(axis='x', rotation=45)

# Ajustar el diseño
fig.tight_layout()
plt.show()
```
