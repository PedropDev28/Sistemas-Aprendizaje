# Función para la creación de arrays:np.linspace

Se utiliza para crear un array NumPy que contiene una secuencia de números espaciados de manera uniforme en un rango específico.

A diferencia de np.arange, donde especificas el valor de inicio, el valor de finalización y el paso, en np.linspace defines el valor de inicio, el valor de finalización y el **número de elementos (o puntos) que deseas entre esos dos valores**.

`numpy.linspace(start, stop, num=50, endpoint=True, dtype=None)`

* start: El valor inicial del rango (inclusivo).
* stop: El valor final del rango. Puede ser inclusivo o exclusivo, dependiendo de endpoint.
* num (opcional): La cantidad de valores que deseas en el array. Por defecto es 50.
* endpoint (opcional): Si es True, el valor de stop estará incluido en el array; si es False, no se incluye. El valor por defecto es True.
* dtype (opcional): El tipo de datos del array resultante. Si no se especifica, NumPy lo deduce automáticamente.

`````python
np.linspace(0, 1, num=10)
`````

`````python
linspace_arr = np.linspace(0, 1, 4)  # Crea un array de 4 valores equidistantes entre 0 y 1
linspace_arr
`````

**np.linspace** es especialmente útil en gráficos, donde necesitas una cantidad fija de puntos igualmente espaciados para generar una curva suave. También es útil en simulaciones y cálculos matemáticos que requieren divisiones precisas de un intervalo.

<h3 id="Matrices">Matrices<a class="anchor-link" href="#Matrices"></a></h3><p>Los vectores son arreglos de una sola dimensión. Las matrices son arreglos de dos dimensiones; generalmente a la primera dimensión se la llama la de las <em>filas</em>, mientras que a la otra se la llama la de las <em>columnas</em>.</p>
<p>Por ende, para crearlas con <code>np.array</code>, necesitamos no una lista de valores, sino una lista de valores <em>por cada fila</em>, o sea, una <em>lista de listas</em>.</p>
<p>Del mismo modo, para crearlas con <code>np.zeros</code> o <code>np.ones</code>, vamos a necesitar una <strong>tupla</strong> con <strong>dos</strong> elementos, uno por cada dimensión.</p>


`````python


print("- Matriz creada con una lista de listas:")
lista_de_listas=[ [1  ,-4], 
                  [12 , 3], 
                  [7.2, 5]]
matriz = np.array(lista_de_listas)
print(matriz)


print("- Matriz creada con np.zeros:")
dimensiones=(2,3)
matriz_ceros = np.zeros(dimensiones)
print(matriz_ceros)


print("- Matriz creada con np.ones:")
dimensiones=(3,2)
matriz_unos = np.ones(dimensiones)
print(matriz_unos)

#también podemos usar np.copy para copiar una matriz 
print("- Copia de la matriz creada con np.ones:")
matriz_unos_copia = np.copy(matriz_unos)
print(matriz_unos_copia)

`````

# Cambio de forma de un array
En NumPy, puedes cambiar la forma de un array existente utilizando el método **reshape** o la función **resize**. 

Esto te permite reorganizar los elementos del array en una nueva forma sin modificar los datos

<h3>#array.resize</h3>
El método resize en NumPy permite cambiar la forma de un array sin alterar sus datos. 

Cambia el tamaño del array en su lugar, es decir, **modifica el array original**. También puede cambiar el número de elementos del array, rellenando con ceros si aumentas su tamaño, o truncando si reduces el número de elementos.

````python
array.resize(new_shape, refcheck=True)

````


* new_shape: La nueva forma que quieres darle al array. Debe ser una tupla que contenga el número de elementos en cada dimensión.
* refcheck (opcional): Por defecto es True. Si es True, NumPy verificará si el array está siendo referenciado en otra parte de tu código, y lanzará un error si intentas cambiar su tamaño. Si es False, ignora esta verificación.

`````python
# array 2D
a = np.arange(12)
print ("Vector original: ", a)
a.resize(4,4) #filas x columnas
print ("Matriz 4x4: ", a)
`````

<h3>#array.reshape</h3>
Se utiliza para cambiar la forma (shape) de un array NumPy existente sin cambiar los datos que contiene.

A diferencia de resize, reshape crea una nueva vista del array con la forma especificada, **sin afectar el array original.**

nuevo_array = array.reshape(new_shape)

* array: El array NumPy que deseas redimensionar.
* new_shape: La nueva forma que deseas para el array, especificada como una tupla de números enteros.

*Reglas para usar reshape*

La cantidad de elementos en el array original y el array resultante deben coincidir. 

Por ejemplo, si el array tiene 12 elementos, puedes reorganizarlo en formas como 3x4, 2x6, etc., pero no 3x5.

Puedes usar -1 para que NumPy calcule automáticamente el tamaño de una de las dimensiones 
en función del número total de elementos.

`````python
a = np.array([1, 2, 3, 4, 5, 6])
b = a.reshape((2, 3))  # Cambia a una forma de 2x3
print(b)

print ()
`````

`````python
a = np.arange(12)
b=a.reshape(2,6)
a, b
`````

`````python
# Se puede usar -1 para que NumPy calcule la dimensión:
a = np.array([1, 2, 3, 4, 5, 6])
b = a.reshape((3, -1))  # NumPy calcula la segunda dimensión
print(b)
`````

¿Convertir un array multidimensional a un array unidimensional:?

`````python
a = np.array([[1, 2, 3], [4, 5, 6]])
b = a.flatten()  # Convierte la matriz a un array unidimensional
#otra forma
c = a.reshape(-1)

#otra forma con resize
a.resize(6)

print(a)
print(b)
print(c)
`````

Diferencias entre <code>reshape</code> y <code>resize</code></h3><ul><li><code>reshape</code> crea una <strong>nueva vista</strong> del array original (sin modificar el array existente), mientras que <code>resize</code> modifica el array <strong>in situ</strong>.</li><li><code>reshape</code> no puede cambiar el número total de elementos, solo la disposición de los mismos. En cambio, <code>resize</code> puede cambiar el tamaño del array agregando ceros o truncando elementos.</li>
