<font color="orange"><strong><h5>LINE PLOT (Gráfico de línea)</strong></h5></font>

Un gráfico de línea, también conocido como "line plot", es uno de los tipos más simples y comunes de gráficos utilizados en Matplotlib. 

Este tipo de gráfico es útil para visualizar la relación entre dos variables **continuas** mediante una línea que conecta los puntos de datos.

**plt.plot()** es una función fundamental en la biblioteca Matplotlib que se utiliza para crear gráficos de líneas, aunque también puede utilizarse para representar otros tipos de gráficos, dependiendo de los datos proporcionados.

Esta función genera un gráfico a partir de los datos de las listas (o arrays) de valores proporcionadas para el eje x y el eje y.


*plt.plot(x, y, 'resto de configuraciones', ...)*
Los parámetros básicos son x e y, que son los valores para los ejes horizontal (x) y vertical (y) respectivamente.

plt.plot() acepta una amplia variedad de parámetros opcionales para personalizar la apariencia del gráfico, como:


1.    *color*: Especifica el color de la línea o puntos.
2.    *linestyle*: Define el estilo de la línea ('-', '--', ':', etc.).
3.    *marker*: Define el marcador utilizado para cada punto en el gráfico.
4.    *label*: Etiqueta para la leyenda del gráfico.
5.    *linewidth*: Ancho de la línea.
6.    *alpha*: Define la transparencia de la línea.


  Entre otros parámetros que pueden personalizar la apariencia del gráfico.

Empezamos por algo sencillo, edad frente a salario

```python
edad = [24,29,34,39,44,49,54,59,64]
salario = [13,18,21,24,28,29,30,32,28]
```

```python
plt.plot(edad, salario)
plt.show()
```

<strong><h5> *Ampliando el gráfico*</strong></h5> 

Podemos añadir más líneas a este gráfico, por ejemplo, si vamos a observar el salario medio respecto a la edad en otro país.

Además, añadimos más funciones:

*plt.legend():*
Esta línea activa la leyenda en el gráfico.


*plt.legend(loc="lower right"):*
Esta línea posiciona la leyenda en el área inferior derecha del gráfico.
Puedes modificar este parámetro para posicionar la leyenda en otros lugares, como "upper left", "upper right", "lower left", entre otros.

*plt.savefig('PlotLine.png'):*
Esta línea guarda el gráfico creado como una imagen en formato PNG en el directorio actual (o en la ubicación especificada).

```python
#España
edad = [24,29,34,39,44,49,54,59,64]
salario = [13,18,21,24,28,29,30,32,28]

#UK
edadUK = [24,29,34,39,44,49,54,59,64]
salarioUK = [18,23,29,30,32,35,33,32,30]

plt.plot(edad,salario,color='green',marker="v",linestyle='solid', label='España')
plt.plot(edadUK,salarioUK,color='red',linestyle='solid', label='UK')

plt.title("Salario medio por rango de edad por países")
plt.xlabel("Rango de edades")
plt.ylabel("Miles de € (k)")
plt.legend()#activamos leyenda
plt.legend(loc="lower right") #posicionamos leyenda

plt.savefig('PlotLine.png') #guardamos el gráfico
plt.show()

```

<font color="orange"><strong><h5>BAR PLOT (Diagrama de barras)</strong></h5></font>
El diagrama de barras, conocido como "bar plot" en inglés, es una representación gráfica que muestra la relación entre variables **categóricas** y sus valores numéricos asociados. Este tipo de gráfico es ideal para comparar diferentes categorías.

En Matplotlib, puedes crear un diagrama de barras utilizando la función ***plt.bar().***


```python
# Datos
producciones = ["La ballena", "Volver", "La Casa de Papel", "Vis a Vis", "El oso"]
num_premios = [3, 5, 8, 4, 15]  # Números de premios o reconocimientos para cada producción

# Crear el diagrama de barras
plt.bar(producciones, num_premios, color='lightgreen')

plt.show()
```

```python
# Personalizar el gráfico
plt.title('Premios')
plt.xlabel('Producciones')
plt.ylabel('Números de premios')


# Crear el diagrama de barras
plt.bar(producciones, num_premios, color='green')
plt.show()
```


Más propiedades para los diagramas de barras:





```python
#podemos pintar cada barra también en un color distinto
colores = ["green", "red", "blue", "yellow", "purple"]
plt.bar(producciones, num_premios, color=colores)
plt.title('Premios')
plt.xlabel('Producciones')
plt.ylabel('Números de premios')
plt.show()
```

```python
#¿ Qué cambia?

plt.barh(producciones, num_premios, color=colores)
plt.title('Premios')
plt.xlabel('Producciones')
plt.ylabel('Números de premios')
plt.show()
```

```python
# Definimos la figura y los ejes del gráfico con Matplotlib(# Crear la figura y los ejes)
fig, axes = plt.subplots()
# Dibujamos las barras con las notas a partir del diccionario
axes.bar(producciones, num_premios, color = 'orange')
# Devolvemos un objeto con los ejes y las barras que contienen
plt.show()
```

En Matplotlib, yticks es una función que se utiliza para establecer la ubicación y las etiquetas de las marcas en el eje vertical (eje y) de un gráfico.
Estos "ticks" son los marcadores a lo largo del eje y que representan valores numéricos.



```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# Crear el gráfico de dispersión
plt.plot(x, y)

# Personalizar las marcas y las etiquetas del eje y
plt.yticks(np.arange(0, 41, 5))

#plt.yticks([10, 20, 30, 40], ['Bajo', 'Medio', 'Alto', 'Altísimo'])
plt.grid() # activamos la cuadrícula
# Mostrar el gráfico resultante
plt.show()
```

<font color="bluelight"><strong><h5>Algunas definiciones importantes:</strong></h5></font>



Los componentes principales de Matplotlib son:

  1. **Figure (Figura)**:  Es el contenedor principal, el lienzo.

  2. **Axes (Ejes)**: El Área del Gráfico. Representa un área de la figura donde se dibuja un gráfico.
Puede incluir títulos, etiquetas de ejes, marcas y leyendas. Cada figura puede contener uno o múltiples objetos axes.

  3. **AxesSubplot:** Subgráficos dentro de una Figura. Se utiliza cuando creas múltiples áreas de gráficos dentro de una figura 

```python
# Crear datos
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

fig, ax = plt.subplots()
ax.plot(x, y)

# Crear figura y eje, especificando el tamaño
fig, ax = plt.subplots(figsize=(3,3))  # Ancho: 3 pulgadas, Alto: 3 pulgadas
ax.set_title("Tamaño Personalizado")
ax.plot(x, y)


plt.show()
```

<font color="green"><h3><strong>SubPlot </strong></h3></font>

Los subplots, o subgráficos, en Matplotlib permiten crear múltiples gráficos dentro de una misma figura.


Esta funcionalidad es útil para comparar diferentes conjuntos de datos o visualizar múltiples gráficos en una misma ventana.

La función *plt.subplots()* se utiliza para crear una matriz de subgráficos en una única figura, permitiendo especificar el número de filas y columnas para organizar los subplots.

*subplot(m, n)*, donde m y n son el tamaño de una matriz que se va a formar.

La variable fig representa la figura completa, 
y axes es una matriz (2x2 en este caso) que contiene los ejes individuales de cada subplot

```python
# Crear subplots con 2 filas y 2 columnas
fig, axes = plt.subplots(nrows=2, ncols=2)
```

```python
# Generar datos para los subplots
# vamos a usar la funcion de numpy linspace: crea un vector de 100 valores, comenzando desde 0 y terminando en 10
import numpy as np
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# Crear subplots
fig, axes = plt.subplots(figsize=(5, 6), nrows=2, ncols=2)

# Graficar en el primer subplot
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Primer subplot')

# Graficar en el segundo subplot
axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Segundo subplot')

# Graficar en el tercer subplot
axes[1, 0].plot(x, y1)
axes[1, 0].set_title('Tercer subplot')

# Graficar en el cuarto subplot
axes[1, 1].plot(x, y2)
axes[1, 1].set_title('Cuarto subplot')

# Ajustar diseño y mostrar los subplots
plt.show()
```

Podemos trabajar directamente con plt.subplot(), sin recuperar el objeto axes, 

indicando en el tercer parámetro con qué gráfico vamos a trabajar

*plt.subplot(nrows, ncols, index)*

```python

plt.figure(figsize=(10, 8)) 

plt.subplot(2, 2, 1)
plt.title("Subplot 1")
plt.plot(x, y1, color='blue')
plt.subplot(2, 2, 3)
plt.title("Subplot 2")
plt.plot(x, y1, color='red')
plt.show()
```

<font color="green"><h3><strong>Integración con Pandas </strong></h3></font>

Matplotlib se integra con la librería Pandas, permitiendo dibujar gráficos a partir de los datos de las series y DataFrames de Pandas.


Los datos de las filas y las columnas a representar deben ser datos del dataframe (las columnas o series, en definitiva)
Se utiliza la siguiente instrucción:


*df.plot(x=columnax, y=columnay, kind=tipo, ax=axes)*: 

Dibuja un diagrama del tipo indicado por el parámetro kind, en el marco indicado en el parámetro ax, representando en el eje x la columna del parámetro x, y en el eje y la columna del parámetro y. 


El parámetro kind puede tomar como argumentos 'line' (lineas), 'scatter' (puntos), 'bar' (barras verticales), 'barh' (barras horizontales), 'hist' (histograma), 'box' (cajas), 'density' (densidad), 'area' (area) o 'pie' (sectores). Es posible pasar otros parámetros para indicar el color, el marcador o el estilo de línea como se vio en los apartados anteriores.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos: Horas dedicadas a distintas actividades en un día
data = {
    'Actividad': ['Dormir', 'Trabajar', 'Comer', 'Ejercicio', 'Ocio', 'Otros'],
    'Horas': [7, 8, 2, 1, 4, 2]
}

# Crear un DataFrame
df = pd.DataFrame(data)

# Crear la figura y los ejes para el gráfico
fig, ax = plt.subplots(figsize=(8, 6))

# Gráfico de barras usando df.plot()
df.plot(x='Actividad', y='Horas', kind='bar', ax=ax, color='skyblue', edgecolor='black')

# Personalización del gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos: Horas dedicadas a distintas actividades en un día
data = {
    'Actividad': ['Dormir', 'Trabajar', 'Comer', 'Ejercicio', 'Ocio', 'Otros'],
    'Horas': [7, 8, 2, 1, 4, 2]
}

# Crear un DataFrame
df = pd.DataFrame(data)

# Configurar el gráfico de pastel
fig, ax = plt.subplots(figsize=(8, 8))

# Gráfico de pastel usando df.plot()
df.plot(
    y='Horas', 
    labels=df['Actividad'], 
    kind='pie', 
    ax=ax, 
    autopct='%1.1f%%', 
    colors=['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#C2C2F0', '#FFB3E6'], 

)



# Generar colores aleatorios
random_colors = np.random.rand(len(df), 3)  # Colores RGB aleatorios

# Personalización del gráfico
ax.set_title('Distribución del tiempo en actividades diarias')
ax.set_ylabel('')  # Elimina la etiqueta del eje Y

# Mostrar el gráfico
plt.show()



```

Características del gráfico de pastel:

    Proporción visual: Muestra cómo se distribuyen las horas en las distintas actividades.
    Colores personalizables: Se aplican colores agradables para distinguir las secciones.
    Formato automático: Usa autopct='%1.1f%%' para mostrar el porcentaje de cada actividad.
